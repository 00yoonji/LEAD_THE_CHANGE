```python
import pandas as pd
import folium
import json
```

```python
Deagusg = pd.read_csv('data/서구서구쓰레기2.csv', index_col = 0,encoding='cp949')
Deagusg.head()
```

```python
m = folium.Map(location=[35.8649155,128.5963041],tiles= 'CartoDB positron', zoom_start=10)

```

```python
deagumap = open('data/서구.json', 'r', encoding='utf-8').read()
deagumap = json.loads(deagumap)
folium.GeoJson(deagumap, name='json_data').add_to(m)
```

```python
for i in Deagusg.index:
    sub_lat =  Deagusg.loc[i,'위도']
    sub_long = Deagusg.loc[i,'경도']
    
    #title = Deagusg.loc[i,'OWNER']
    
    #지도에 데이터 찍어서 보여주기
    folium.Circle([sub_lat,sub_long],tooltip = "쓰레기", color='#000000', fill='crimson', radius = 50).add_to(m)
```

```python

m.save('서구.html')
m
```
